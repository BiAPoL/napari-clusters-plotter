```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt


# the order of the results is always: cle, numpy, mapping(with adding layer), mapping (without adding layer)
results_3d_2dviewer = [812.5,968.75,140.625,78.125]
results_4d_2dviewer = [1640.625,2906.25,187.5,46.875]
results_3d_3dviewer = [7406.25,9718.75,8093.75,7734.375]
results_4d_3dviewer = [8109.375,11328.125,8421.875,7703.125]



data = [
    results_3d_2dviewer,
    results_4d_2dviewer,
    results_3d_3dviewer,
    results_4d_3dviewer,
]

labels = [
    '3D-Lab. - 2D Viewer',
    '4D-Lab. - 2D Viewer',
    '3D-Lab. - 3D Viewer',
    '4D-Lab. - 3D Viewer',
]

data_transposed = np.asarray(data).T

methods = ['CLE', 'NUMPY', 'CMAP w. add layer', 'CMAP wo add layer']

from napari_clusters_plotter._utilities import get_nice_colormap
cmap = get_nice_colormap()
```

```python
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
for i in range(4):
    ax.bar(X + i*0.2, data[i], color = cmap[i], width = 0.2, label = labels[i])
ax.legend()
plt.xticks([x+0.3 for x in X],methods)
```
