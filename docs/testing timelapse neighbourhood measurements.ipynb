```python
import napari
viewer = napari.Viewer()
viewer.add_image(image)
viewer.add_labels(labels)
```

```python
napari_image = viewer.layers['image'].data
napari_image.shape
```

```python
napari_labels = viewer.layers['labels'].data
napari_labels.shape
```

```python
regionprops_napari = new_regprops(
    intensity_image=napari_image,
    label_image=napari_labels,
    region_props_source='neighborhood',
    timelapse=True,
)
regionprops_napari
```

```python
from napari.qt.threading import create_worker

def result_of_get_regprops(returned):
    returned.to_csv('test.csv')
    

worker = create_worker(
    new_regprops,
    intensity_image=napari_image,
    label_image=napari_labels,
    region_props_source='neighborhood',
    timelapse=True,
)
worker.returned.connect(result_of_get_regprops)
worker.start()
```
