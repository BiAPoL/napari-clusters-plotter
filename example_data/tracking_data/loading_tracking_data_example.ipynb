```python
import numpy as np
import napari
import pandas as pd
from skimage.io import imread
from napari.utils import nbscreenshot
from napari_clusters_plotter._plotter import PlotterWidget

data = pd.read_csv("tgmm-mini.csv")

tracking_label_image = imread("tgmm-mini.tif")

# For tracks visualization only
tracking_layer_data = pd.read_csv("tgmm-mini-tracks-layer-data.csv").to_numpy()

viewer = napari.Viewer(ndisplay=3)

viewer.add_labels(tracking_label_image, features=data)
viewer.add_tracks(tracking_layer_data) # not used by napari-clusters-plotter
viewer.camera.angles = (0, 10, 150)
```
